<hr><h4>vmin & vmax</h4>

```python
fig,(ax1,ax2) = plt.subplots(1,2,constrained_layout=True, figsize=(8,3))

# subplot 1
img1 = ax1.imshow(im, vmin = 700 , vmax=1000)
ax1.set_title("vmin = 700 , vmax=1000")
cbar = fig.colorbar(img1)
_ = cbar.set_label(label="Elevatiion")

# subplot 2
img2 = ax2.imshow(im, vmin = 800 , vmax=900)
cbar = fig.colorbar(img2)
ax2.set_title("vmin = 800 , vmax=900")
_ = cbar.set_label(label="Elevatiion")
```

<hr><h3>Lognorm</h3>

```python
from matplotlib.colors import LogNorm

# take the 2D fft of our test image and re-order to put 0 frequency in the center
spike = np.abs(np.fft.fftshift(np.fft.fft2(im)))

fig, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True, figsize=(8, 3))

lin_im = ax1.imshow(spike)
_ = fig.colorbar(lin_im, ax=ax1)
_ = ax1.set_title('linear color scale')

log_im = ax2.imshow(spike, norm=LogNorm())
_ = fig.colorbar(log_im, ax=ax2)
_ = ax2.set_title('log color scale')
```

```python
import matplotlib.colors as mcolors
import matplotlib.cm as mcm

cmap = mcm.tab20b
norm = mcolors.BoundaryNorm([1, 3, 4, 5, 7, 9, 12, 15, 20], cmap.N)

data = np.arange(20).reshape((1, 20))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))

top = ax1.imshow(data, cmap=cmap, aspect='auto')
fig.colorbar(top, ax=ax1)
bottom = ax2.imshow(data, norm=norm, cmap=cmap, aspect='auto')
fig.colorbar(bottom, ax=ax2)
```

```python
import matplotlib
import matplotlib as mpl


days = ["Sun", "Mon", "Tue", "Wed",
              "Thu", "Fri", "Sat"]
person = ["Ram", "Sham", "Chhotu",
           "Motu", "Chhutki", "Raju", "Babu Bhaiya"]

work = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 2.0],
                [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 2.0],
                [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 3.0],
                [0.6, 0.0, 0.3, 5.0, 3.1, 1.0, 0.0],
                [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 1.0],
                [1.3, 6.2, 0.0, 0.5, 0.0, 3.2, 5.1],
                [0.1, 2.0, 0.5, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(work, cmap='Reds')

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(person)), labels=person)
ax.set_yticks(np.arange(len(days)), labels=days)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(days)):
    for j in range(len(person)):
        text = ax.text(j, i, work[i][j],
                       ha="center", va="center", color="black")

ax.set_title("Work profile in a week")
fig.tight_layout()
plt.show()
```
