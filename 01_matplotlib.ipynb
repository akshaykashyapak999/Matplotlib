<hr>here np.sort() will sort age (increasing order) that will go into x-axis<hr>

```python
fig, ax = plt.subplots()
# providing data to x, y axis
# plot means 2Dline plot , 
_ = ax.plot(np.sort(data['age']), marker ='o', markersize=1)

_ = ax.set_ylabel("Age") # y label
_ = ax.set_xlabel("Passenger") # x label
_ = ax.set_title("DATASET") # title
```

<hr> <h3><code>multiple axes</code></h3><hr>

```python
fig, (ax1, ax2) = plt.subplots(nrows= 2)
_ = fig.suptitle("Two plots In 2 rows")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols= 2)
_ = fig.suptitle("Two plots in 2 col")
_ = ax1.set_title("AAAAAAAAAA")
_ = ax2.set_title("BBBBBBBBBB")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2, gridspec_kw={'width_ratios': [1, 3]})
_ = fig.suptitle("2 plots with different widths")
plt.show()
```

<div class="alert alert-block alert-info" style="margin-top: 20px"><h3><code> subplots()</code> v/s
<code>subplot()</code></h3></div>

```python
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

x2 = np.random.rand(100)
y2 = np.random.rand(100)

x3 = ['A', 'B', 'C']
y3 = [3, 7, 2]

f = [ 'm','m','f','m','f','f','f','f','m','m','m','m','f','f','f','m','m','m','m','m']
```

<hr>subplots()

```python
# here we will use 2rows 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x1, y1)
axes[0, 1].scatter(x2, y2)
axes[1, 0].bar(x3, y3)
axes[1, 1].hist(f)
plt.show()
```

<hr>subplot(  grid_row , grid_col , position)<br>
subplot(2,2 , 1) => 2X2 grid ; position 1

```python
plt.subplot(2, 2, 1)
plt.plot(x1, y1)

plt.subplot(2, 2, 2)
plt.scatter(x2, y2)

plt.subplot(2, 2, 4)
plt.bar(x3, y3)

plt.subplot(2, 2, 3)
plt.hist(f)

plt.show()
```
