```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
# laoding data
df= pd.read_csv("matplotlib_data/titanic.csv").dropna()
```

<hr><div class="alert alert-block alert-danger" style="margin-top: 20px"><h3>Histogram</h3></div>
<hr><div>
A <code>histogram</code> is a graphical representation of <code>data distribution</code> that uses <code>bars</code><br>
to display the <code>frequency</code> of numerical data within specific ranges, called <code>bins</code>. It provides a <br>
visual summary of the data's distribution, showing patterns such as central tendency, variability, and skewness.
</div><hr>

```python
fig, ax = plt.subplots()
_ = ax.hist('age',data=df,bins='auto', density=True)
```

```python
# custom bins & color
fig, ax = plt.subplots()
_ = ax.hist('age',data=df,bins=[0,18,25,50,80], density=True, color='gray') 
```

```python
# histtype & color 
# xticks 
fig, ax = plt.subplots()
_ = ax.hist('age',data=df,bins=[0,18,25,50,80], density=True, histtype='step', color="red")
_ = ax.set_xticks([0,18,25,50,80])
```

```python
# edgecolor & hatch
# color in hex 
fig, ax = plt.subplots(figsize=(5,3))
_ = ax.hist('age',data=df,bins=[0,18,25,50,80], density=True, color="#00001a", edgecolor='hotpink', hatch="\\")
_ = ax.set_xticks([0,18,25,50,80])
```

```python
# rwidth & density
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(8,3))
_ = ax1.hist('age',data=df,bins=[0,18,25,50,80], density=True, color="#1166aa",
            rwidth=0.9, edgecolor='black', hatch=".")

_ = ax2.hist('age',data=df,bins=[0,18,25,50,80], density=False, color="#660099",
            rwidth=0.9, edgecolor='black', hatch="\\")
#xticks
_ = ax1.set_xticks([0,18,25,50,80])
_ = ax2.set_xticks([0,18,25,50,80])

plt.show(ax1,ax2)
```

<div>When <code>density=True</code>, each bar's <code>height is scaled</code> so that the total <code>area</code> under the histogram <code>equals 1</code>.<br>
    This normalization adjusts the height of each bar based on the number of data points in each bin<br>
    relative to the total number of data points.<br><br>
Without normalization <code>(density=False)</code>, the height of each bar directly corresponds<br>
    to the <code>number</code> of data points falling into that bin.<br><hr>
<code>NOTE:</code> If <code>bins="auto"</code>or <code>uniform bin size</code> then no scaling, Scaling only happening due to <code>non-uniform</code> bin size</div><hr>

```python
# rwidth & density
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(8,3))
_ = ax1.hist('age',data=df,bins="auto", density=True, color="#1166aa",
            rwidth=0.9, edgecolor='black', hatch=".")

_ = ax2.hist('age',data=df,bins="auto", density=False, color="#660099",
            rwidth=0.9, edgecolor='black', hatch="\\")
#xticks
_ = ax1.set_xticks([0,18,25,50,80])
_ = ax2.set_xticks([0,18,25,50,80])

ax1.set(xlabel="Age", ylabel="Proportion", title="Density=True")
ax2.set(xlabel="Age", ylabel="No of passenger", title="Density=False")

plt.show(ax1,ax2)
```

To prevent subplots from colliding we can do one of two things

<hr> Let's Try muliple histgram in one plot

```python
fig, ax = plt.subplots(figsize=(5,3))

# groupby to seperate male female 
for label, gdf in df.groupby('sex'):
    ax.hist('age', data=gdf, bins="auto", label=label.upper(),alpha=0.5) # upper to capitalize

# labels, title,legend
_ = ax.set(xlabel="Age", ylabel="No of people", title="No of people v/s Age (groubed by sex)")
_= ax.legend()
```
