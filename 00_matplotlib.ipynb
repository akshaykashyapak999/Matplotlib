Let's try `multiple` in histogram

```python
male_marks = [80,78,79,85,96,95,75,84,88,99,96,85,84,82,76,73,74,76,84,95,90,84,83,80,86,87,87]
female_marks = [90,78,96,85,96,95,80,88,90,99,96,93,90,85,79,80,79,76,84,95,90,84,83,85,90,88,97]
```

```python
plt.figure(figsize=(5,3)) # we use to define size to plot

plt.hist(male_marks, bins=5, alpha=0.5, label='Male Marks')
plt.hist(female_marks, bins=5, alpha=0.5, label='Female Marks')

plt.title('Histogram of Male and Female Marks')
plt.legend()
```

## add KDE through seaborn on the same plot

```python
import seaborn as sns  # Import seaborn for KDE plot

# Histograms
plt.hist(male_marks, bins=10, alpha=0.5, label='Male Marks', density=True)
plt.hist(female_marks, bins=10, alpha=0.5, label='Female Marks', density=True)

# KDE plots
sns.kdeplot(male_marks, color='blue', linestyle='-', linewidth=2, label='Male Marks KDE')
sns.kdeplot(female_marks, color='orange', linestyle='-', linewidth=2, label='Female Marks KDE')

plt.xlabel('Marks')
plt.ylabel('Density')
plt.title('Histogram with KDE for Male and Female Marks')
plt.legend()

plt.show()
```

<hr><div class="alert alert-block alert-danger" style="margin-top: 20px">
    <h3>Pie</h3>
</div><hr>
<div>
<code>Pie</code> plot, or pie chart, is a circular statistical graphic that is divided into <code>slices</code> to illustrate<br>
    numerical proportions. Each slice represents a <code>percentage</code> or proportion of a whole, making it useful for<br>
    visualizing relative sizes or <code>distributions</code> of categorical data.</div>

```python
label = ['A-expense', 'B-expense', 'C-expense', 'D-expense']
vals = [1400,600,400,250]
```

```python
plt.pie(vals, labels=label,autopct='%0.1f%%')

# autopct (autopercentage) , to add percenatge label
plt.show()
```

`explode` always take value and `multiply` with `radius` provided also take `-ve` value

```python
plt.pie(vals, labels=label,autopct='%0.1f%%', shadow= True, explode=[0,0.1,2,-0.2])
plt.show()
```

```python
import pandas as pd
import seaborn as sns

df= sns.load_dataset('titanic')

plt.pie(x=df['survived'].value_counts(), labels=['Died', 'Survived'],
           colors=['grey', 'green'], autopct='%1.2f%%', startangle=90)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Survival Distribution')  # Optional title
plt.show()
```
